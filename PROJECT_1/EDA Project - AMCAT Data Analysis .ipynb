```python
from fuzzywuzzy import process

def correct_spelling_errors(target_word = '', choices = [], threshold = 80):
    match, score = process.extractOne(target_word, choices)
    if score >= threshold:
        return match
    else:
        return target_word
```

```python
Q1 = num_columns['salary'].quantile(0.25)
Q3 = num_columns['salary'].quantile(0.75)

IQR = Q3 - Q1

l1 = (Q1 * 1.5 - IQR)
l2 = (Q3 * 1.5 + IQR)
```
